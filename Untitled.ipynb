Затем объявим необходимые функции, которые помогут в подготовке датасета и в визуализации данных.

Это будут
* Функции с использованием *Natural Language Toolkit*, которые будут использоваться для лемматизации;
* Функция для вывода пар <стобец, коэффициент> для логистической регрессии;
* Функция для вывода красивой матрицы ошибок.

```python
import nltk
from nltk.corpus import wordnet
from nltk.corpus import stopwords


# набор функций для лемматизация
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize_text(text):
    temp = [lemmatizer.lemmatize(w.lower(), get_wordnet_pos(w.lower())) for w in w_tokenizer.tokenize(text)]
    return (' '.join(map(str, temp))).strip()


# Вывод топа коэффициентов логистической регрессии
def get_logistic_regr_top(feature_names, coefs):
    top = pd.DataFrame({'Word':feature_names, 'Weight':coefs})
    print(top.sort_values("Weight"))  
    
    
# Вывод матрицы ошибок
def show_confusion_matrix(matrix):
    plt.title('Матрица ошибок')
    sns.heatmap(pd.DataFrame(matrix), annot=True, cmap="viridis" ,fmt='g')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    
    
```

 ##  Стоп слова, лемматизация, выбор признаков

В качестве стоп слов будем использовать *символы пунктуации*, *url'ы* и *набор стоп слов* для английского языка из *nltk*.

```python
from string import punctuation

stop_words = list(punctuation) + ['http', 'co'] + stopwords.words('english')
```

К столбцу **"text"** применим созданную выше функцию для лемматизации. 

Для русского языка, согласно википедии:
* для существительных — именительный падеж, единственное число;
* для прилагательных — именительный падеж, единственное число, мужской род;
* для глаголов, причастий, деепричастий — глагол в инфинитиве (неопределённой форме) несовершенного вида.

```python
df['text'] = df.text.apply(lemmatize_text)
```

```python
df.head()
```

Выбор признаков

```python
#X_columns = ['text', "keyword"]
X_columns = ['text']
y_columns = ['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[X_columns], df[y_columns], random_state=42)
```

# Мешок слов

Первым способом для кодирования предложений будем использовать модель мешка слов. Т.е. исходный столбец с тектом твитта превратится в *M* столбцов, где *M* -- колечество уникальных N-грамм во всех текстах. В каждом столбце будет стоять 0 или 1 в зависимости встречается ли данная N-грамма в тексте.

Опытным путем приходим к выводу, что будем использовать **униграммы**.

## Логистическая регрессия

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report

N = 1

# Готовим трансформер
transformer = ColumnTransformer([
    ("t", CountVectorizer(ngram_range=(N, N), stop_words=stop_words), "text"),
    #("kk", OneHotEncoder(handle_unknown="ignore"), ["keyword"]),
], remainder="passthrough")

# Преобразуем тренировочный набор данных
x_train_prepared = transformer.fit_transform(x_train)

# Создаем логистическую регрессию с дефолтными (почти) параметрами
# и обучаем модель
clf = LogisticRegression(random_state=42, max_iter=1000)
clf.fit(x_train_prepared, y_train.values.ravel())

# Проверяем все на тестовых данных
x_test_prepared = transformer.transform(x_test)
pred = clf.predict(x_test_prepared)

# Отчет
print(classification_report(y_test, pred))
```

```python
show_confusion_matrix(confusion_matrix(y_test, pred))
```

Согласно отчету, модель одинаково хорошо группирует твитты в оба класса.

Рассморим коэффициенты логистической регрессии. 

* Если коэффициент близок к нулю, то он не влияет на принятие решения;
* Если положительный, то влияет в сторону одного класса;
* Иначе в сторону другого. 

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): # чтобы вывести все
  display(get_logistic_regr_top(np.array(transformer.get_feature_names()), clf.coef_[0]))
```

## Случайный лес

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report

N = 1

transformer = ColumnTransformer([
    ("t", CountVectorizer(ngram_range=(N, N)), "text"),
], remainder="passthrough")
x_train_prepared = transformer.fit_transform(x_train)


clf = RandomForestClassifier(n_estimators=50, max_depth=20, random_state=42)
clf.fit(x_train_prepared, y_train.values.ravel())
x_test_prepared = transformer.transform(x_test)
pred = clf.predict(x_test_prepared)
print(classification_report(y_test, pred))

show_confusion_matrix(confusion_matrix(y_test, pred))
```

В случае со случайным лесом не удалось подобрать оптимальные настройки. Так что имеем модель, которая большую часть тестовой выборки распознала как не относящуюся к реальности, но то реальное, что она угадала - было угадано хорошо.

# TF-IDF

Открывок о TF-IDF.

`TfidfVectorizer` делает то же, что и `CountVectorizer`, но в качестве значений выдает **tf-idf** каждого слова.

Как считается tf-idf:

**TF (term frequency)** – относительная частотность слова в документе:
$$ TF(t,d) = \frac{n_{t}}{\sum_k n_{k}} $$

**IDF (inverse document frequency)** – обратная частота документов, в которых есть это слово:
$$ IDF(t, D) = \mbox{log} \frac{|D|}{|{d : t \in d}|} $$

Перемножаем их:
$$TFIDF(t, d, D) = TF(t,d) \times IDF(i, D)$$

Сакральный смысл: если слово часто встречается в одном документе, но в целом по корпусу встречается в небольшом 
количестве документов, у него высокий TF-IDF.

## Логистическая регриссия

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report

N = 1

transformer = ColumnTransformer([
    ("t", TfidfVectorizer(ngram_range=(N, N), stop_words=stop_words), "text")
], remainder="passthrough")
x_train_prepared = transformer.fit_transform(x_train)


clf = LogisticRegression(random_state=42)
clf.fit(x_train_prepared, y_train.values.ravel())

x_test_prepared = transformer.transform(x_test)
pred = clf.predict(x_test_prepared)
print(classification_report(y_test, pred))
```

```python
show_confusion_matrix(confusion_matrix(y_test, pred))
```

В случае с TF-IDF + логистической регрессией мы получили модель схожую с мешком слов + лог. регрессией.

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): # чтобы вывести все
  display(get_logistic_regr_top(np.array(transformer.get_feature_names()), clf.coef_[0]))
```
